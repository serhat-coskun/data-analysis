```python
import pandas as pd
```

```python
#load data
df = pd.read_csv("pokemon_data.csv")
#Excel alternative
xl = pd.read_excel("pokemon_data.xlsx")
#Tab seperated file
tab = pd.read_csv("pokemon_data.txt", delimiter="\t")
```

```python
# Same data we will follow with df
print(df.head(3)) #head-tail pair to access first or last n rows
print(xl.head(3))
print(tab.head(3))
```

```python
#Properties of df

print(df.columns) #name of the columns
print(df.dtypes) #types of columns
print(df.shape) #(rows, columns)
```

```python
# Fast summary function for numeric variables(columns)
df.describe()
```

```python
# Access a column, 2 alternatives
print(df['Name'].head(5))
print(df.Name.head(5))
# Access multiple columns, note must give list of column names, not tuple
print(df[['Name', 'HP']].head(5))
```

```python
# Access rows
print(type(df.iloc[0])) #if only one row is accessed returns Series
print(type(df.iloc[1:5])) #more than one returns dataframe
# Possible to use booleans, slices, lists, constants, predicates
#1 slice
print(df.iloc[0:1])
#2 constant
print(df.iloc[0]) 
#3 lists
print(df.iloc[[0]])
#4 booleans
print(df.iloc[list(df.HP < 50)])
#5 predicate
print(df.loc[lambda x: (x.HP < 50)])

# iloc, loc also enables indexing for columns
print(df.iloc[[0], 1])
```

```python
# Iterate over rows using an iterator
# Might be usefull when no built-in functionality is avaliable in pd
for index, row in df.iterrows():
    print(index, row.Name)
```

```python
# Drop an row, or column
# Caution not in-place
xl.drop(index=df[5:798].index, inplace=True)
print(xl)

xl.drop(columns=['Name', 'Type 1'], inplace=True)
print(xl)
```

```python
# Compare iloc vs loc
# Loc uses row labelsi or booleans, better with predicates(callables)
# iloc uses integer row posisiton
```

```python
#Sorting
df.sort_values(by=["Type 1","HP"], ascending=False)
```

```python
#Filtering data 1
#Chooose any col that satisfies conditions
new_df = df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison')]
# Set drop = true, because it does not discard old index automatically
new_df.reset_index(inplace = True, drop=True)
new_df
```

```python
#Filtering 2
import re

df.loc[df["Name"].str.contains("meGa", regex = True, flags = re.I)]
df.loc[df["Name"].str.contains("pi[a-z]*", regex = True, flags = re.I)]
```

```python
# Conditional changes
df.loc[df["Type 1"] == "Fire", "Type 1"] = "Fuss"
# Conditional change on multiple col at the same time
df.loc[df["Total"] > 600, ["Generation", "Legendary"]] = [-1,True]

df
```

```python
# Aggregate stats
# Refresh df df = pd.read_csv("pokemon_data.csv")


df.groupby(["Type 1"]).mean().sort_values(by ="Defense" ,ascending=False)

df.groupby(["Type 1"]).mean().sort_values(by ="Attack" ,ascending=False)
```

```python
df['Total'] = df.iloc[:,4:10 ].sum(axis = 1)
df.groupby("Type 1").mean().sort_values("Total", ascending=False)
```

#### Working with large amounts of data

```python
new_df = pd.DataFrame(columns=df.columns)
for df in pd.read_csv("pokemon_data.txt", chunksize=5):
    results = df.groupby('Type 1').count()
    new_df = pd.concat([new_df, results])
```
