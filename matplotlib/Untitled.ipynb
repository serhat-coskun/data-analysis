```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import matplotlib.style
import matplotlib as mpl
```

```python
gas = pd.read_csv("gas_prices.csv")
fifa = pd.read_csv("fifa_data.csv")
```

Line Graph

```python


plt.figure(figsize = (15,5))
title = plt.title("Gas prices($) vs Years")
plt.setp(title, color='r')         #set the color of title to red

for name in gas.columns:
    if name == "Year":
        continue
    plt.plot(gas["Year"], gas[name], linestyle="dashed", label = name)

plt.xlabel("Years")
plt.ylabel("Price in dollars")
plt.xticks(gas.Year[::3])
plt.legend()
```
